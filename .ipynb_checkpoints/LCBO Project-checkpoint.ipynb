```python
import flask
from  splinter import  Browser 
from bs4 import BeautifulSoup 
import pandas as pd
import requests
import time
from selenium.webdriver.common.by import By
browser = Browser('chrome')
```

```python
url = "https://www.lcbo.com/en/products#t=clp-products&sort=relevancy&layout=card"
browser.visit(url)
browser.links.find_by_text("English").click()
browser.find_by_id("btn-cookie-allow").click()
# browser.driver.find_element(by=By.XPATH, value='//button[@class="my_store_close_button"]').click()

```

```python
i = 0
while i <= 200:
    try:
        time.sleep(1)
        browser.find_by_id("loadMore").click()
        print(i)
        i+=1
        if i == 100:
            html = browser.html
            soup = BeautifulSoup(html,'html.parser')
            all_links = soup.find_all("div",class_ = 'coveo-product-items' )
            link1 = []
            for l1 in all_links:
                re = l1.a['href']
                link1.append(re)
            print(len(link1))
            df = pd.DataFrame(link1)
            df.to_csv(r'C:\Users\Play\Documents\GitHub\LCBO-Project\LCBO links5k.csv')
        elif i == 187:
            html = browser.html
            soup = BeautifulSoup(html,'html.parser')
            all_links = soup.find_all("div",class_ = 'coveo-product-items' )
            link1 = []
            for l1 in all_links:
                re = l1.a['href']
                link1.append(re)
            print(len(link1))
            df = pd.DataFrame(link1)
            df.to_csv(r'C:\Users\Play\Documents\GitHub\LCBO-Project\LCBO links9k.csv')
            
            
    except:
        break
        
        
        
        
    
    
    
```
